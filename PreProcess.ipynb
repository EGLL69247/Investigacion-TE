# Script para descargar archivos desde un link 

##### Los modulos os y time solo son para comprobar cuanto tiempo se tarda en descargar archivos. El módulo ThreadPool te permite ejecutar varios subprocesos utilizando el grupo. 

```python
import os 
import requests 
from time import time
from multiprocessing.pool import ThreadPool 
```

```python
def url_response(url):
    path,url = url
    r = requests.get(url,stream=True)
    with open(path,'wb') as f:
        for ch in r:
            f.write(ch)
```
